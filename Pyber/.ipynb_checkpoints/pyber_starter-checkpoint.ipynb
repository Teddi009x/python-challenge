```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset

cc_data = pd.merge(city_data, ride_data, how="left", on=["city", "city"])

# Display the data table for preview

cc_data.head()
```

```python
#   *Average Fare ($) Per City
city_fare_avg = cc_data.groupby(['city'])['fare'].mean()
city_fare_avg.head()
```

```python
city_fare_sum = cc_data.groupby(['city'])['fare'].sum()
city_fare_sum.head()
```

```python
#  Total Number of Rides Per City
rides_per_cc = cc_data.groupby(['city'])['ride_id'].count()
rides_per_cc.head()
```

```python
#  Total Number of Drivers Per City
driver_per_cc = cc_data.groupby(['city'])['ride_id'].count()
driver_per_cc.head()
```

```python
#city_types=city_data.set_index(["type"])["city"].count()
city_types=city_data.set_index(["city"])["type"]

city_types.head()
```

```python
city_type_grp = city_data.groupby(['type'])['city']
city_type_grp
```

```python
PyBer_df = pd.DataFrame({'Avg Fare':city_fare_avg,
                  'Rides Per City':rides_per_cc,
                  'Drivers Per City':driver_per_cc,
                  'City Type':city_types})

PyBer_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_grp = PyBer_df.loc[PyBer_df['City Type'] == 'Urban']
Suburban_grp = PyBer_df.loc[PyBer_df['City Type'] == 'Suburban']
Rural_grp = PyBer_df.loc[PyBer_df['City Type'] == 'Rural']

#add city group type count; then insert after y variable ex: s= driver_group_city

# Build the scatter plots for each city types
plotx1 = urban_grp.plot(kind='scatter',x='Rides Per City', y='Avg Fare',s=PyBer_df['Drivers Per City']*10,color='red',label = 'Urban')
plotx2 = Suburban_grp.plot(kind='scatter',x='Rides Per City', y='Avg Fare',s=PyBer_df['Drivers Per City']*10,color='blue',label = 'Suburban')
plotx3 = Rural_grp.plot(kind='scatter',x='Rides Per City', y='Avg Fare',s=PyBer_df['Drivers Per City']*10,color='orange',label = 'Rural')


# Incorporate the other graph properties
plotx1 = urban_grp.plot(kind='scatter',x='Rides Per City', y='Avg Fare',s=PyBer_df['Drivers Per City']*10,color='red',label = 'Urban')
plotx2 = Suburban_grp.plot(kind='scatter',x='Rides Per City', y='Avg Fare',s=PyBer_df['Drivers Per City']*10,color='blue',label = 'Suburban', ax=plotx1)
plotx3 = Rural_grp.plot(kind='scatter',x='Rides Per City', y='Avg Fare',s=PyBer_df['Drivers Per City']*10,color='orange',label = 'Rural',ax=plotx1)

# Create a legend

plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.legend(title = 'City Types')
plt.grid(True)

# Incorporate a text label regarding circle size
#  .annotate('...Additional information...')


```

```python
# Save Figure
plt.tight_layout()
plt.savefig("pyber_ride_sharings.png")
plt.show()
```
